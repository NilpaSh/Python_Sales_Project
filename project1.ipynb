```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
comments["comment_text"]
```

```python
sia.polarity_scores("MY FAN . attendance")
```

```python
sia.polarity_scores("MY FAN . attendance")['compound']
```

```python
sia.lexicon
```

```python

def analyzer_comment_sentiment(comment):
    compound = sia.polarity_scores(comment)['compound']
    
    if compound > 0.05:
            label = "Positive"
            insight = "the comment express positive emotion or approval"
        
    elif compound < 0.05:
            label = "Negative"
            insight = "the comment express dis-satifaction emotion or negative"
            
    else:
            label = "Netural"
            insight = "the comment express informational emotion"
            
    return{
        "label":label,
        "score" : compound,
        "insight" : insight
    }
```

```python
analyzer_comment_sentiment("MY FAN . attendance")
```
